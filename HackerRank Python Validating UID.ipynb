# HackerRank Python Validing UID

```python
# we need to work with regular expressions for this problem.
import re

for _ in range(int(input())):
    u = ''.join(sorted(input()))
    
    try:
        # The "assert" keyword is used when debugging code. 
        # The assert keyword lets us test if a condition in 
        # your code returns True, if not, Python will 
        # raise an AssertionError. 
        
        # It must contain at least 2 uppercase 
        # English alphabet characters.
        assert re.search(r'[A-Z]{2}', u)
        
        # It must contian at least 3 digits (0-9)
        assert re.search(r'\d\d\d', u)
        
        # No character should repeat.
        assert not re.search(r'[^a-zA-Z0-9]', u)
        assert not re.search(r'(.)\1', u)
        
        # there must be exactly 10 characters in a valid UID
        assert len(u) == 10
        
    except:
        print('Invalid')
    else:
        print('Valid')
```
