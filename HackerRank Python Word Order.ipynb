# HackerRank Python Word Order

```python
#OrderedDict library is great for this use case:
from collections import OrderedDict

# instantitate OrderedDict() instance:
words = OrderedDict()

# for the range of the number of words we will input:
for _ in range(int(input())):
    
    # we input each word
    word = input()
    
    # add one to the count each time we come across that word
    words.setdefault(word, 0)
    words[word] += 1

# first line: print number of DISTINCT words from input
print(len(words))

# second line: print number of occurrences for each distinct word
print(*words.values())
```
