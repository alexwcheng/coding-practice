Using the notebook from yesterday, please construct a call to find all the tracks that have a unit price greater than .99. Have the call return the track name, the artist name, and the unit price.
This could also been expressed as tracks.name, artists.name, and invoice_items.unitprice. The call you will make will use JOIN and WHERE statements.
Respond in this thread with the SQL query that you wrote.

```python
#this was your original answer to this warmup...
import pandas as pd #import pandas
import sqlite3 #import sqlite3

conn = sqlite3.connect('/Users/alexandercheng/Desktop/code/warmups/SQLite Warmup/chinook.db') #establish connection with file location
cur = conn.cursor() #load cursor

cur.execute("""
SELECT t.name AS Track_Name, ar.Name AS Artist, ii.unitprice AS Price -- select all the items we want in our final output.
FROM artists AS ar
JOIN albums AS al
USING(ArtistId) -- JOIN artists and albums using (ArtistId) which is common between artists and albums tables.
JOIN tracks AS t
USING(AlbumId) -- JOIN tracks with (AlbumId) which is common between tracks and albums tables.
JOIN invoice_items AS ii
USING(TrackId) -- JOIN invoice_items with (TrackId) which is common bewteen invoice_items and tracks tables.
WHERE ii.unitprice > .99
;""")

df = pd.DataFrame(cur.fetchall()) #assign fetchall() to the df dataframe
df.columns = [x[0] for x in cur.description] #label columns
df #return my dataframe
```
