## Routes

```python
pd.read_sql('''
SELECT * FROM routes
''', con)
```

```python
# Which countries have the highest amount of INACTIVE airports?

pd.read_sql('''
SELECT country, COUNT(*) as airline_count FROM airlines
WHERE active = 'N'
GROUP BY country
ORDER BY (airline_count) DESC
LIMIT 10
''', con)
```

```python
# Which countries have the highest amount of ACTIVE airports?

pd.read_sql('''
SELECT country, COUNT(*) as airline_count FROM airlines
WHERE active = 'Y'
GROUP BY country
ORDER BY (airline_count) DESC
LIMIT 10
''', con)
```

```python
# Which timezone has the most airports?

pd.read_sql('''
SELECT name, timezone, COUNT(*) as timezone_count FROM airports
GROUP BY timezone
ORDER BY (timezone_count) DESC
LIMIT 10
''', con)
```

```python
pd.read_sql('''
SELECT * FROM airlines al
INNER JOIN airports ap
USING ("id")
INNER JOIN routes rt
    ON al.id = rt.airline_id
LIMIT 1000
''', con)
```

```python
#What is the range of stops made?
pd.read_sql('''
SELECT DISTINCT stops FROM routes
''', con)
```
