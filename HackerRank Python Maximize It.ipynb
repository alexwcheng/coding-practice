# HackerRank Python Maximize It!

```python
from itertools import product

# provide input value of K (no. of rows) and M(modulus) 
# convert both values to integer using map() function
K,M = map(int,input().split())

# Create an empty list
# loop K (no. of row ) times
nums = []
for _ in range(K):
    
# use map and split function to convert row input into list of int.
# use [1:] to slice out the first number of each row...
# which is actually the count of items in that row, which we don't need.
    row = map(int,input().split()[1:])
    
# we square and find the remainder(modulus) 
# after dividing the squared number by M for each number in the row,
# we append that to the "nums" empty list.
    nums.append(map(lambda x:x**2%M, row))

# The greatest number of this "nums" list is our answer. 
# Use the "max" function for finding the biggest number.
print(max(map(lambda x: sum(x)%M, product(*nums))))
```

**In the above example, "388" is the greatest "remainder" possible from squaring one number from each list, and using a modulus of "500".**
