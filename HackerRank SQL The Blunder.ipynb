# HackerRank SQL The Blunder

```python
# pseudocode
# provide "+1" for the round, because we are rounding up no matter what.
SELECT ROUND(AVG(SALARY) - AVG(REPLACE(SALARY,'0',''))+1) FROM EMPLOYEES;
```

```python
# CEIL() returns smallest integer that is bigger or equal to a number.
# Can use CEIL() instead of ROUND(+1)
SELECT CEIL(AVG(Salary)-AVG(REPLACE(Salary,'0',''))) FROM EMPLOYEES;
```
