# HackerRank SQL Binary Tree Nodes

```python
#pseudocode
SELECT N, 
IF(P IS NULL,'Root',IF((SELECT COUNT(*) FROM BST WHERE P=B.N)>0,'Inner','Leaf')) 
FROM BST AS B 
ORDER BY N;
```

```python
# Alternative Solution:
select N, 
case when P is NULL then 'Root' when N in (select P from BST) 
then 'Inner' else 'Leaf' end as Node 
from BST 
order by N;
```
