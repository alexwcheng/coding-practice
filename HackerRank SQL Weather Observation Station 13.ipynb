# HackerRank SQL Weather Observation Station 13

```python
# pseudocode
SELECT TRUNCATE(SUM(LAT_N),4)
FROM STATION
WHERE LAT_N BETWEEN 38.7880 AND LAT_N 137.2345;
```

```python
# alternative answer instead of using "BETWEEN"
SELECT TRUNCATE(SUM(LAT_N),4)
FROM STATION
WHERE LAT_N>38.7880 AND LAT_N<137.2345;
```
