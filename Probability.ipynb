Out of a group of people, 20% speak English and Spanish. 60% of the people in the group speak English. What is the probability that a person from this group speaks Spanish, given that they speak English? Given the below data raw_data create a scatterplot of preTestScore and postTestScore where the size of each point is determined by age and the color is determined by sex. Include a title, x-axis label, and y-axis label.

Let a person speaking English be Event A. P(A) = 0.6
Let a person speaking Spanish be Event B. P(B) = ???
Let a person speaking English and Spanish be Event AB. P(AB) = 0.2

P(AB) = P(B|A) / P(A)
Solve for P(B|A)

P(B|A) = 0.2 / 0.6
P(B|A) = .33

______

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73], 
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (6,3))
plt.scatter(raw_data['preTestScore'], raw_data['postTestScore'], 
            marker='o', c = raw_data['female'], s = raw_data['age'])
plt.title('Warmup Graph')
plt.xlabel('preTestScores')
plt.ylabel('postTestScores')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x =  raw_data['preTestScore'], y = raw_data['postTestScore'], hue = raw_data['female'], size = raw_data['age'], sizes = (20,200))
plt.xlabel('PreTest Scores')
plt.ylabel('PostTest Scores')
```
