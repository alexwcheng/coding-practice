# HackerRank Python Detect HTML Tags, Attributes, and Attribute Values

```python
# library to help us extract text from HTML code
from html.parser import HTMLParser

# create a function to extract the tag name and attributes
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(tag)
        for attr in attrs:
            print("->", attr[0], ">", attr[1])

# instantiate MyHTMLParser() as "parser"
parser = MyHTMLParser()

# feed MyHTMLParser the HTML code in the input box
for i in range(int(input())):
    parser.feed(input())
```
